```python
''' 3. A dictionary stores usernames and passwords.
Write a python program to validate user login.'''

users = {"admin": '1234', "student": "abcd"}
user_name = input("Username: ")
pw = input("Passwords: ")

if users.get(user_name) == pw:
  print("Login sucessfull")
else:
  print("Invalid username or password")
```

```python
'''4. A dictionary stores student names and marks.
Write a python program to find the student with highest marks '''


marks = {"A": 85, "B": 92, "C": 78}
topper = max(marks, key=marks.get)
print("Topper:", topper)
print("Marks:", marks[topper])
```
